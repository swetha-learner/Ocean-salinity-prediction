##  1. Plot a scatter plot of temp vs salinity. What is the relationship between them?

```python
temp = df['T_degC']
sal = df['Salnty']

import matplotlib.pyplot as plt
plt.figure(dpi=120)
plt.title('Ocean Salinity vs Temperature')
plt.xlabel("Temperature")
plt.ylabel("Salinity")
plt.scatter(temp,sal,color='m',marker = '.') #fmt
plt.axis([2,15,32,35])  #x1,x2,y1,y2
plt.grid(alpha = 0.5)
plt.show()
```
